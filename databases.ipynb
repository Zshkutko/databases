```python
def add_game(genre_id, developer_id, difficulty_id, game_name):
    cur.execute(f'''
        INSERT INTO game (genre_id, developer_id, difficulty_id, game_name) VALUES 
        ('{genre_id}', '{developer_id}', '{difficulty_id}','{game_name}')
        ''')
    con.commit()
    
    cur.execute(f'''
    SELECT id FROM game WHERE game_name = '{game_name}'
    ''')
    
    
    _id = cur.fetchall()[0]['id']
    
    cur.execute(f'''
        INSERT INTO game2publisher (game_id, publisher_id) VALUES 
        ('{_id}', '{random.randint(0,25)}')
        ''')
    con.commit()
    
    cur.execute(f'''
    SELECT id FROM game2publisher
    ''')
    
    _id = cur.fetchall()[-1]['id']
    
    
    
    cur.execute(f'''
        INSERT INTO game2platform (game2publisher_id, platform_id, release_date) VALUES 
        ('{_id}', '{random.randint(0,5)}', '{random.randint(1990, 2021)}')
        ''')
    con.commit()

    cur.execute('''
    SELECT * FROM game
    ''')
    

    return cur.fetchall()
```

```python
def delete_game(by):
    if type(by) == int:
        cur.execute(f'''
            DELETE FROM game WHERE id = {by}''')
        con.commit()

        cur.execute('''
        SELECT * FROM game
        ''')
    if type(by) == str:
        cur.execute(f'''
            DELETE FROM game WHERE game_name = '{by}' ''')
        con.commit()

        cur.execute('''
        SELECT * FROM game
        ''')
    

    return cur.fetchall()
```

```python
def change_game_name(idx, new_name):
    cur.execute(f'''
        UPDATE game SET game_name = '{new_name}' WHERE id = '{idx}' ''')
    con.commit()

    cur.execute('''
    SELECT * FROM game
    ''')
    

    return cur.fetchall()
```

```python
def gen_game(f, s, t, p):
    return ' '.join(random.sample(f, 1) + random.sample(s, 1) + random.sample(t, 1) + random.sample(p, 1))

def gen_publisher(f, s):
    return ' '.join(random.sample(f, 1) + random.sample(s, 1))

def gen_developer(f, s):
    return ' '.join(random.sample(f, 1) + random.sample(s, 1))

def random_date(start,l):
    current = start
    while l >= 0:
        curr = current + datetime.timedelta(minutes=randrange(60))
        yield curr
        l-=1
```

# SELECT с 3+ таблицами

Топ-20 издательств по количеству игроков на PS4 и PS5 в Европе

```python
cur.execute('''
SELECT players, platform_name, release_date, region_name, publisher_name 
FROM region2players, platform, game2platform, game2publisher, region, publisher
WHERE region2players.game2platform_id = game2platform.id
AND game2platform.platform_id = platform.id
AND game2platform.game2publisher_id = game2publisher.id
AND region2players.region_id = region.id
AND game2publisher.publisher_id = publisher.id
AND platform_name IN ('PS4', 'PS5')
AND region_name = 'Europe'
ORDER BY players DESC
LIMIT 20
''')

data = cur.fetchall()
pd.DataFrame(data)
```

Игры с жанром RPG и разработчиками с 'Blue' в начале никнейма

```python
cur.execute('''
SELECT game_name FROM game, genre, developer
WHERE game.genre_id = genre.id
AND game.developer_id = developer.id
AND genre_name = 'RPG'
AND developer_name REGEXP 'Blue*'
''')

data = cur.fetchall()
pd.DataFrame(data)
```

# Вложенные SELECT-запросы

Максимальное количество игроков на регион

```python
cur.execute('''
SELECT MAX(total_players) AS record_players_num FROM
(SELECT region_name, SUM(players) AS total_players FROM region2players, region
WHERE region2players.region_id = region.id
GROUP BY region_name) as t1
''')
            
cur.fetchone()
```

Издательства игр на PS4 и PS5 в Европе, количество игроков которых превышает среднее

```python
cur.execute('''
SELECT AVG(players) FROM region2players
''')
            
cur.fetchone()
```

```python
cur.execute('''
SELECT players, platform_name, release_date, publisher_name FROM
(SELECT players, platform_name, release_date, region_name, publisher_name 
FROM region2players, platform, game2platform, game2publisher, region, publisher
WHERE region2players.game2platform_id = game2platform.id
AND game2platform.platform_id = platform.id
AND game2platform.game2publisher_id = game2publisher.id
AND region2players.region_id = region.id
AND game2publisher.publisher_id = publisher.id
AND platform_name IN ('PS4', 'PS5')
AND region_name = 'Europe'
ORDER BY players) as t2, 
(SELECT AVG(players) AS avg_players FROM region2players) as t3
WHERE t2.players > t3.avg_players
''')

data = cur.fetchall()
pd.DataFrame(data)
```

# Последовательный JOIN

Общая таблица с информацией по играм

```python
cur.execute('''
SELECT game_name, developer_name, publisher_name, genre_name FROM game2publisher
JOIN publisher ON publisher_id = publisher.id
JOIN game ON game_id = game.id
JOIN developer ON developer_id = developer.id
JOIN genre ON genre_id = genre.id
''')

data = cur.fetchall()
pd.DataFrame(data)
```

# AGGREGATE + JOIN + SORT

Количество игр каждого уровня сложности

```python
cur.execute('''
SELECT difficulty_name, COUNT(*) AS difficulty_count
FROM game
JOIN difficulty ON difficulty_id = difficulty.id
GROUP BY difficulty_id ORDER BY difficulty_count DESC
''')

data = cur.fetchall()
pd.DataFrame(data)
```

# AGGREGATE + SORT

Рейтинг регионов по суммарному количеству игроков

```python
cur.execute('''
SELECT region_name, SUM(players) AS total_players_sum FROM region2players, region
WHERE region2players.region_id = region.id
GROUP BY region_name ORDER BY total_players_sum DESC
''')

data = cur.fetchall()
pd.DataFrame(data)
```

# SORT

Общее количество игроков на всех платформах по названиям игр

(последовательные джойны с сортировкой)

```python
cur.execute('''
SELECT game_name, SUM(players) AS total_players_sum FROM region2players
JOIN game2platform ON game2platform_id = game2platform.id
JOIN game2publisher ON game2publisher_id = game2publisher.id
JOIN game ON game_id = game.id
JOIN platform ON platform_id = platform.id
GROUP BY game_name ORDER BY total_players_sum DESC
''')

data = cur.fetchall()
pd.DataFrame(data)
```
